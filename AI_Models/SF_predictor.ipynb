```python
# Final Model: Random Forest for SF Prediction
# Uses all features except ToA (leakage) and device_id (identifier)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    ConfusionMatrixDisplay
)

# 1) Define target
y_final = df['SF']

# 2) Define features
# Remove:
# - SF: target column
# - toa: removed because it caused data leakage
# - device_id: identifier, not a meaningful numeric feature

X_final = df.drop(columns=[
    'SF',
    'toa',
    'device_id'
], errors='ignore')

# Keep numeric columns only, as Random Forest cannot use text values directly
X_final = X_final.select_dtypes(include=np.number)

# 3) Train-test split
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(
    X_final,
    y_final,
    test_size=0.20,
    random_state=42,
    stratify=y_final
)

# 4) Build the final Random Forest model
final_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

# 5) Train model
final_model.fit(X_train_final, y_train_final)

# 6) Predict
y_pred_final = final_model.predict(X_test_final)

# 7) Evaluate exact accuracy
exact_accuracy_final = accuracy_score(y_test_final, y_pred_final)

# 8) Evaluate practical accuracy within ±1 SF
within_1_accuracy_final = np.mean(
    np.abs(y_test_final.to_numpy() - y_pred_final) <= 1
)

print("=" * 50)
print("FINAL RANDOM FOREST MODEL RESULTS")
print("=" * 50)
print(f"Exact Accuracy: {exact_accuracy_final:.4f}")
print(f"Within ±1 SF Accuracy: {within_1_accuracy_final:.4f}")

print("\nClassification Report:")
print(classification_report(y_test_final, y_pred_final))

# 9) Feature importance
feature_importance_final = pd.Series(
    final_model.feature_importances_,
    index=X_final.columns
).sort_values(ascending=False)

print("\nFeature Importance:")
print(feature_importance_final)

# 10) Confusion Matrix
ConfusionMatrixDisplay.from_predictions(
    y_test_final,
    y_pred_final,
    cmap='Blues'
)

plt.title("Final Random Forest Confusion Matrix")
plt.show()
```

```python
# Final Deployable Model: SF Predictor

import numpy as np
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier

y_deploy = df['SF']

X_deploy = df.drop(columns=[
    'SF',          # Target
    'toa',         # Removed because of data leakage
    'device_id'    # Identifier, not a useful prediction feature
], errors='ignore')

X_deploy = X_deploy.select_dtypes(include=np.number).copy()

final_sf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

final_sf_model.fit(X_deploy, y_deploy)

model_package = {
    'model': final_sf_model,
    'feature_columns': X_deploy.columns.tolist(),
    'target': 'SF',
    'removed_columns': ['toa', 'device_id']
}

joblib.dump(model_package, 'final_sf_predictor.pkl')

print("Final SF prediction model saved successfully.")
print("Model file: final_sf_predictor.pkl")
print("Number of features:", len(X_deploy.columns))
print("Features used:", X_deploy.columns.tolist())
```
